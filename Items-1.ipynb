```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

db_filename = 'dinofunworld-1.db'

conn =sqlite3.connect(db_filename)
c = conn.cursor()

#################################################### 饼图
c.execute("select attraction, count(*) as visit_number from checkin where type ='actual' and attraction in (select attractionID from attraction where Category like '%Thrill Rides%') group by attraction")
counts = c.fetchall()
manuStats = pd.DataFrame.from_records(counts, columns=['attraction', 'visit_number'])
##print(manuStats)
plt.pie(manuStats['visit_number'], labels = manuStats['attraction'], shadow=False)
plt.axis('equal')
plt.show()
```
